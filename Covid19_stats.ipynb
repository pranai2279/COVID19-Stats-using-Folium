```python
import pandas as pd
import numpy as np
import folium as fo
import branca
%matplotlib inline
```

```python
map = fo.Map()
```

```python
df = pd.read_csv('covid19.csv')
df.head()
```

```python
df1 = df.tail(35)
```

```python
df1.Latitude[df1['Name of State / UT'] == 'Dadra and Nagar Haveli and Daman and Diu'] = 20.4283
df1.Longitude[df1['Name of State / UT'] == 'Dadra and Nagar Haveli and Daman and Diu'] = 72.8397
df1.Latitude[df1['Name of State / UT']=='Telangana'] = 18.1124
df1.Longitude[df1['Name of State / UT']=='Telangana'] = 79.0193
```

```python
lat = list(df1.Latitude)
lon = list(df1.Longitude)
nam = list(df1['Name of State / UT'])
tt = list(df1['Total Confirmed cases'])
de = list(df1.Death)
cu = list(df1['Cured/Discharged/Migrated'])
```

```python
co = fo.FeatureGroup(name='covid map')
```

```python
def fancy_html(row):
    i = row
    
    name = df1['Name of State / UT'].iloc[i]
    tot = df1['Total Confirmed cases'].iloc[i]
    dea = df1.Death.iloc[i]
    cur = df1['Cured/Discharged/Migrated'].iloc[i]
    Date = df1['Date'].iloc[i]
    left_col_colour = "#2A799C"
    right_col_colour = "#C5DCE7"
    
    html = """<!DOCTYPE html>
<html>

<head>
<h4 style="margin-bottom:0"; width="300px">{}</h4>""".format(Date) + """

</head>
    <table style="height: 126px; width: 300px;">
<tbody>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Name of state</span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{}</td>""".format(name) + """
</tr>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Total confirmed cases</span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{}</td>""".format(tot) + """
</tr>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Deaths</span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{}</td>""".format(dea) + """
</tr>
<tr>
<td style="background-color: """+ left_col_colour +""";"><span style="color: #ffffff;">Cured/Discharged</span></td>
<td style="width: 200px;background-color: """+ right_col_colour +""";">{}</td>""".format(cur) + """
</tr>
</tbody>
</table>
</html>
"""
    return html
```

```python
for la,lo,nm,tc,i in zip(lat,lon,nam,tt,range(0,len(df1))):
    html = fancy_html(i)
    iframe = branca.element.IFrame(html=html,width=400,height=300)
    popup = fo.Popup(iframe,parse_html=True)
    co.add_child(fo.Marker(location=[la,lo], popup=popup, icon=fo.Icon(color='red'))).add_to(map)
```
